### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head(15)
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

# Urban
#urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]

#urban_ride_count = urban_cities_df.count()["ride_id"]
#print(urban_ride_count)

#Suburban
#suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]

#suburban_ride_count = suburban_cities_df.count()["ride_id"]
#suburban_ride_count

#Rural

#rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]

#rural_ride_count = rural_cities_df.count()["ride_id"]
#rural_ride_count

# Woah that's a complicated way to do it, let's make it easy

rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
rides_by_type


```

```python
# 2. Get the total drivers for each city type

#urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]

#drivers_by_type = pyber_data_df.groupby(["type"]).count()["driver_count"]
#drivers_by_type

#pyber data dataframe contains duplicates
#to get rid of dupes, use city data

drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
drivers_by_type


```

```python
#  3. Get the total amount of fares for each city type

sum_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
mean_fares_by_type = pyber_data_df.groupby(["type"]).mean()["fare"]
mean_fares_by_type
```

```python
# 5. Get the average fare per driver for each city type. 

average_driver_fare = pyber_data_df.groupby(["type"]).sum()["fare"] / city_data_df.groupby(["type"]).sum()["driver_count"] 

average_driver_fare
```

```python
#  6. Create a PyBer summary DataFrame. 

pyber_summary_df = pd.DataFrame({"Drivers per City Type" : drivers_by_type, "Number of Rides per City Type" : rides_by_type, "Total Fares per City Type" : sum_fares_by_type, "Average Fare per City Type" : mean_fares_by_type, "Average Fare per Driver per City Type" : average_driver_fare})
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.

pyber_summary_df["Total Fares per City Type"] = pyber_summary_df["Total Fares per City Type"].map("${:.2f}".format)



pyber_summary_df["Average Fare per City Type"] = pyber_summary_df["Average Fare per City Type"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver per City Type"] = pyber_summary_df["Average Fare per Driver per City Type"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame


pyber_data_df.head(15)
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

fare_by_date = pyber_data_df.groupby(["date"]).sum()["fare"]
fare_by_date

```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

fare_by_date = fare_by_date.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

fares_index_date = pd.pivot_table(pyber_data_df, values = "fare", index = "date", columns = "type")
fares_index_date.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

#create a variable for the date range to obtain the indecies
#then loc in the dataframe with those indicies

date_range = (fares_index_date.index > '2019-01-01') & (fares_index_date.index <= '2019-04-29')

print(fares_index_date.loc[date_range])

fares_index_date = fares_index_date.loc[date_range]
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

fares_index_date.index = pd.to_datetime(fares_index_date.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()

fares_index_date.info()

#fares_index_date.index.info() is more correct I think but it gives an error

#that error actually has the dtype as DatetimeIndex which is what it is

#an n-dimensional array with datetimes in it

```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.


weekly_fares = fares_index_date.resample('W').sum()
weekly_fares
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

x_axis = weekly_fares.index


fig, ax = plt.subplots(figsize = (30,12))
ax.plot(weekly_fares)
ax.set_ylim([0,2750])
#ax.set_xticks(weekly_fares.index[::5])
#ax.set_xlim(["2019-01-06", "2019-05-12"])
ax.set_ylabel("Fares ($ USD)", size = 28)
ax.set_xlabel("Month", size = 28)
ax.set_title("Total Fare by City Type", size = 34)

ax.set_xticks(x_axis[::5])
ax.set_xticklabels(['January', "Feb", "March", "April"])
ax.tick_params(axis='both', which='major', labelsize=25)

#fig.legend()


plt.tight_layout()
#ax.locator_params(axis = x_axis, nbins = 4)
#plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=2))

#import matplotlib.dates as mdates
#x_axis.set_major_locator(locator)

fig.legend(weekly_fares,loc = "center", title = "type", fontsize = "large", title_fontsize = "large", bbox_to_anchor=(0.27, 0.25, 0.5, 0.5))

plt.savefig("CityTypeFares.png")

```
